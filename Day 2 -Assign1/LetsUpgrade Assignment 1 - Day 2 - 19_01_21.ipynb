Write a Python Program for factorial of a number.

```
num = input("Enter a number: ")
num = int(num)

fact = 1

if (num == 0):
    fact = 1 
while(num>0):
    fact=fact*num
    num=num-1
print("Factorial of the number is: ")
print(fact)
```

Find the sum of n numbers by using the while loop

```
num = input("Enter a number: ")
num = int(num)

if num<0:
    print("Oops!")
    print("Please enter a number greater than zero")
else:
    Sum = 0
    while num>0:
        Sum = Sum + num
        num = num - 1
    print("Sum of the numbers is: ",Sum)
```

Write a Python program to print all Prime numbers in an Interval.

```
low = input("Enter a lower interval: ")
low = int(low)
upp = input("Enter a higher interval: ")
upp = int(upp)

print("Prime numbers between", low, "and", upp, "are:")

for num in range(low, upp + 1):
   
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```
